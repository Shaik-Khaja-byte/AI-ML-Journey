```python
# continue - used to skip the current iteration and move to the next
i = 1
while(i <= 10):
    if (i % 3 == 0):
        i += 1
        continue
    print(i)
    i += 1
```

```python
# print only the odd numbers using continue keyword
i = 1
while(i <= 10):
    if (i % 2 == 0):
        i += 1
        continue
    print(i)
    i += 1
```

```python
# WAF that takes 3 inputs and calculates the average of three numbers
def average(a, b, c):
    return (a+b+c)/3

print(average(1,2,3))
print(average(15,10,5))
```

```python
# you can assign default values to the parameters, when any arguments are missing the function can use these values
# function(non_default_params, default_params) => syntax of defining default params

def average(a, b, c=1):
    return (a+b+c)/3

print(average(1,2,3))
print(average(28,1))
```

```python
# lambda functions : small anonymous functions defined using lambda keyword
# syntax - lambda (parameters) : (expression)
# now this lambda function returns single expression_value
# not used for complex calcualtions
# used in high order functions

sum = lambda a,b : a+b
print(sum(1,2))

avg = lambda a,b,c : (a+b+c)/3
print(avg(1,2,3))
```

```python
# WAF to print factorial of n
def fact(n):
    if (n==0 or n==1):
        return 1
    return n * fact(n-1)

print(fact(5))
print(fact(4))
```

```python
# can be written using loop too
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

n = int(input())
print(factorial(n))
```

```python
# create a function simple_calculator(a, b, operation)

def simple_calculator(a, b, operation):
    if operation == '+':
        return a+b
    elif operation == '-':
        return a-b
    elif operation == '*':
        return a*b
    elif operation == '/':
        return a/b
    else:
        return None

print(simple_calculator(1,2,'+'))
print(simple_calculator(1,2,'-'))
print(simple_calculator(1,2,'*'))
print(simple_calculator(1,2,'/'))
print(simple_calculator(1,2,'@'))
```

```python
# check whether a number is prime or not
def is_prime(n):
    for i in range(2, n):
        if n%i == 0:
            return False
    return True
        
n = int(input())
print(is_prime(n))
```

```python
# guessing game
code = 12
while True:
    guess = int(input())
    if guess == code:
        print("Correct!")
        break
    elif guess > code:
        print("Too High")
    else:
        print("Too Low")
```
