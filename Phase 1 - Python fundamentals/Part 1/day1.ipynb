```python
# variables
# these are considered as the containers that store the data
# names of these variables are called as identifiers
# there are certain rules to follow while naming a variable
# 1. it can contain A-Z and a-z (upper case and lowercase letters are different)
# 2. it can contain numbers, but should not start with them
# 3. it can start with an _, no special symbols are allowed

name = "Shaik Khaja"
_age = 20
# 1percentage = 99.99  | can't start with a number
percentage = 99.99

print(name, _age, percentage)
```

```python
# datatypes

# integer - +ve, -ve, 0
num = -99
print(num, "is : ", type(num))

# decimal - floating point numbers
dec = 6.9
print(dec, "is : ", type(dec))

# string - chars enclosed within quotes
str = "Spiderman"
print(str, "is : ", type(str))

# boolean - True or False
var = True
print(var, "is : ", type(var))

# none - when you don't wanna assingn anything to a variable
x = None
print(x, "is : ", type(x))
```
