```python
# variables
# these are considered as the containers that store the data
# names of these variables are called as identifiers
# there are certain rules to follow while naming a variable
# 1. it can contain A-Z and a-z (upper case and lowercase letters are different)
# 2. it can contain numbers, but should not start with them
# 3. it can start with an _, no special symbols are allowed

name = "Shaik Khaja"
_age = 20
# 1percentage = 99.99  | can't start with a number
percentage = 99.99

print(name, _age, percentage)
```

```python
# datatypes

# integer - +ve, -ve, 0
num = -99
print(num, "is : ", type(num))

# decimal - floating point numbers
dec = 6.9
print(dec, "is : ", type(dec))

# string - chars enclosed within quotes
str = "Spiderman"
print(str, "is : ", type(str))

# boolean - True or False
var = True
print(var, "is : ", type(var))

# none - when you don't wanna assingn anything to a variable
x = None
print(x, "is : ", type(x))
```

```python
# arithmetic operators

a = 10
b = 5

print(a+b)
print(a-b)
print(a*b)
print(a/b) 
print(a%b) # modulo - remainder calculation
print(a**b) # exponent operator - calculates a^b
```

```python
# relational operators - whose output is either True or False
a = 10
b = 4

print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b)
```

```python
# assignment operator '=' , if a = b means b value is stored in a
a = 5
a += 1
print(a)
a -= 1
print(a)
a *= 2
print(a)
a /= 2
print(a)
a **= 2
print(a)
a %= 2
print(a)
```

```python
# logical operators - combining two or more relational operators
# and - binary multiplication (1 * 1 => 1) i.e., only if both the conditions are true the output is true
# or - binary addition (0 + 0 => 0) i.e., only if both the conditions are false the output is false
# not - reverse the output

print((5 > 3) and (4 < 0)) # False
print((1 != 1) or (1 >= 1)) # True
print((not(0)))
```

```python
# operator precedence
# () > ** > *,/,% > +,- > relational operators > logical operators (not, and, or)
# if operators have same precedence solve from left to right

ans = ((5+1)*2/2**2)
print(ans) # 3
```

```python
# type conversion - implicit by the interpreter
# type casting - explicit by the developer using int(), float() functions
# all these happen only with the compatible types

ans1 = 10/2
ans2 = int(10/2)

print(ans1, type(ans1))
print(ans2, type(ans2))

val = "123"
lav = int(val)

print(val, type(val))
print(lav, type(lav))
```
