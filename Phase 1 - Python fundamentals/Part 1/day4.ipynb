```python
# classes and objects
# class - blueprint of an object
# object - instance of a class
# instance means they are the main thing - which actually takes up the memory, class doesn't occupy space

class Student:
    subject = "Python"
    college = "BITM"
    year = "4th"

stud1 = Student() # creating the object
print(stud1)

# accessing the attributes
print(stud1.subject, stud1.college, stud1.year)
```

```python
l = [1,2,3]
print(type(l)) # <class, list> we are able to create list directly because the blueprint is already written 
s = set() # creating an object for a set, it's blue print is also written
print(type(s)) # <class set>
```

```python
# a class conatins attributes(properties) and methods(behaviours)
# methods in classes
# __init__ method (constructior) - which is used to initialize the objects
class Student:
    def __init__(self, name, cgpa): # self is a compulsary parameter which refers to the current instance
        self.name = name # current instances name(in the memory) has the value name(which you passes as argument)
        self.cgpa = cgpa

s1 = Student("Shaik Khaja", 9.41)
s2 = Student("Rahul B", 8.7)
s3 = Student("Saraswath", 9.7)
s4 = Student("Shreyas", 8.1)

print(s1.name, s1.cgpa)
print(s2.name, s2.cgpa)
print(s3.name, s3.cgpa)
print(s4.name, s4.cgpa)
```

```python
# we can have other instance methods but it must have a cumpulsory instance parameter - self
class Student:
    def __init__(self, name, cgpa): 
        self.name = name
        self.cgpa = cgpa

    def get_cgpa(self):
        return self.cgpa
    

s1 = Student("Akanksh", 9.6)
s2 = Student("Sundeep", 8.5)

print(s1.get_cgpa())
print(s2.get_cgpa())
```

```python
# two types of constructors in python - default(only self) and parameterized
# in python only one constructor in a class, even if multiple exists... last one is considered
# per class only one init method.

class Student:
    def __init__(self): # default
        print("initializing objects...")

    def __init__(self, name, cgpa): # parameterized
        self.name = name
        self.cgpa = cgpa

    def get_cgpa(self):
        return self.cgpa
    
s1 = Student("Akanksh", 9.6)
s2 = Student("Sundeep", 8.5)

print(s1.get_cgpa())
print(s2.get_cgpa())
```
