```python
# classes and objects
# class - blueprint of an object
# object - instance of a class
# instance means they are the main thing - which actually takes up the memory, class doesn't occupy space

class Student:
    subject = "Python"
    college = "BITM"
    year = "4th"

stud1 = Student() # creating the object
print(stud1)

# accessing the attributes
print(stud1.subject, stud1.college, stud1.year)
```

```python
# a class conatins attributes(properties) and methods(behaviours)
# methods in classes
# __init__ method (constructior) - which is used to initialize the objects
class Student:
    def __init__(self, name, cgpa): # self is a compulsary parameter which refers to the current instance
        self.name = name # current instances name(in the memory) has the value name(which you passes as argument)
        self.cgpa = cgpa

s1 = Student("Shaik Khaja", 9.41)
s2 = Student("Rahul B", 8.7)
s3 = Student("Saraswath", 9.7)
s4 = Student("Shreyas", 8.1)

print(s1.name, s1.cgpa)
print(s2.name, s2.cgpa)
print(s3.name, s3.cgpa)
print(s4.name, s4.cgpa)
```

```python
# we can have other instance methods but it must have a cumpulsory instance parameter - self
class Student:
    def __init__(self, name, cgpa): 
        self.name = name
        self.cgpa = cgpa

    def get_cgpa(self):
        return self.cgpa
    

s1 = Student("Akanksh", 9.6)
s2 = Student("Sundeep", 8.5)

print(s1.get_cgpa())
print(s2.get_cgpa())
```

```python
# two types of constructors in python - default(only self) and parameterized
# in python only one constructor in a class, even if multiple exists... last one is considered
# per class only one init method.

class Student:
    def __init__(self): # default
        print("initializing objects...")

    def __init__(self, name, cgpa): # parameterized
        self.name = name
        self.cgpa = cgpa

    def get_cgpa(self):
        return self.cgpa
    
s1 = Student("Akanksh", 9.6)
s2 = Student("Sundeep", 8.5)

print(s1.get_cgpa())
print(s2.get_cgpa())
```

```python
# attributes are of two types - class (common for all objects, e.g., clg_name) and instance (vary with objects, e.g., name)
class Student:
    clg_name = "BITM" # class attribute
    PI = 3.1

    def __init__(self, name, cgpa): 
        self.name = name # instance attribute
        self.cgpa = cgpa
        self.PI = 3.14

s1 = Student("Ajay", 7.9)
print(s1.name)
print(s1.clg_name) # we can access both class and instance attribute using object
print(Student.clg_name) # but with the help of class name only class attribute can be accessed

# if both class and instance attributes have same name
print(s1.PI) # at that time instance has higher priority when compared to class
print(Student.PI) # as you explicitly called using class name
```

```python
# 3 types of methods
# type 1 - instance methods
# has self as a compulsary attribute

class Laptop:
    storage_type = "ssd"

    def __init__(self, RAM, storage):
        self.RAM = RAM
        self.storage = storage

    # there is an instance method that prints the info of the object
    def get_info(self):
        print(f"Laptop has {self.RAM} RAM & {self.storage} {self.storage_type}")

l1 = Laptop("16gb", "512gb")
l2 = Laptop("8gb", "256gb")

l1.get_info()
l2.get_info()

# interesting thing to notice here is self which is an instance argument can also access the class attributes
```

```python
# class methods - parameter cls
# can be identifed with a decorator @classmethod - which transforms the behaviour of the method to a class method
# can only access the class attributes, has no knowledge about the instance attributes
# can be accessed with the help of objects also as they have the info about the class

class Laptop:
    storage_type = "ssd"

    def __init__(self, RAM, storage):
        self.RAM = RAM
        self.storage = storage

    @classmethod  # decorator
    def get_storage_type(cls):
        print(f"Laptop has {cls.storage_type} as the storage type")

l1 = Laptop("16gb", "512gb")
l2 = Laptop("8gb", "256gb")

l1.get_storage_type()
Laptop.get_storage_type()
```

```python
# static methods - no cumpolsory parameter, it doesn't use class or instance attributes
# these are just standalone functions in the class with a decorator @staticmethod
# use - sometimes we need some functions and which are logically tied up with the classes so we don't wanna separate them
class Laptop:
    storage_type = "ssd"

    def __init__(self, RAM, storage):
        self.RAM = RAM
        self.storage = storage

    @staticmethod
    def get_discount(price, discount):
        final_amt = price - (price * discount / 100)
        print(f"discounted price is : {final_amt}")

l1 = Laptop("16gb", "512gb")
l2 = Laptop("8gb", "256gb")

l1.get_discount(40_000, 10)
```

```python
class Product:
    count = 0 # class attribute

    def __init__(self, prod, price):
        self.prod = prod
        self.price = price
        Product.count += 1

    def get_info(self): # instance method
        print(f"Product : {self.prod}, Price : {self.price}")

    @classmethod
    def count_prod(cls):
        print(F"Total products created : {cls.count}")

    @staticmethod
    def get_discount(price, discount):
        final_amt = price - (price * discount / 100)
        print(f"discounted price is : {final_amt}")

p1 = Product("Phone", 10_000)
p2 = Product("Laptop", 50_000)
p3 = Product("Charger", 3500)

Product.count_prod()

p1.get_info()

p2.get_discount(p2.price, 12)
```

```python
# let's dicuss the four pillars of OOP
# encapsulation - wraping the data andd it's functions into a single unit
# with encapsulation comes the concept of data hiding - public, protected, private
# there is no direct way to make it private, protected, by convention the programmers follow and understand
# _ single underscore means protected - attribute will be accessible but by covention it's a bad practice
# __ means private - if you try to access data mangling occurs (for some extent data is protected)
# now we understand that some things need to be kept private and should not be accessible outside the class
# if at all you want to access it - getter and setter functions (which is rather a more safe way to access)

class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance

    def get_balance(self):
        return self.__balance
    
    def set_balance(self, new_balance):
        self.__balance = new_balance

c1 = BankAccount("Shaik Khaja", 1_00_000)
print(c1.get_balance())
c1.set_balance(2_00_000)
print(c1.get_balance())
# there is a correct way to access the private attribute (objName._className__privateMember)
print(c1._BankAccount__balance)
```

```python
# inheritaance - reusing attributes and methods fron the parent(base) class
class Employee:
    start_time = "10am"
    end_time = "5pm"

    def change_end_time(self, endtime):
        self.end_time = endtime

class Teacher(Employee):
    def __init__(self, subject):
        self.subject = subject

class AdminStaff(Employee):
    def __init__(self, role):
        self.role = role

e1 = Teacher("math")
print(e1.subject, e1.start_time, e1.end_time)
e2 = AdminStaff("Manager")
e2.change_end_time("6pm")
print(e2.role, e2.start_time, e2.end_time)
```

```python
# multilevel inheritance : G -> P -> C
# c class has all the attributes and methods of G and P
class Employee:
    start_time = "10am"
    end_time = "5pm"

class AdminStaff(Employee):
    def __init__(self, role):
        self.role = role

class Accountant(AdminStaff):
    def __init__(self, salary, role):
        super().__init__(role)
        self.salary = salary

acc1 = Accountant(100_000, "CA")
print(acc1.salary, acc1.role, acc1.start_time, acc1.end_time)
```
