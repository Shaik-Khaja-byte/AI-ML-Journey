```python
# string formatting
# 1. using format() 

# normal formatting
a = 1
b = 2
sum = a + b
print("sum of {} and {} is {}".format(a, b, sum))
print("This is the {} of {} function\n".format("result", "format()"))

# index based formatting
print("sum of {1} and {0} is {2}\n".format(a, b, sum))

# value based formatting
print("sum of {x} and {y} is {z}".format(x=10, y=20, z=30)) # useful when you want to reuse the variables
```

```python
# 2. using f-strings - literal string interpolation
a = 10
b = 20
print(f"avg of {a} and {b} is {(a + b)/2}")
```

```python
# lists - mutable sequence of values
marks = [78, 90, 69, 99, 58, "abc", 93.67]
print(type(marks))

marks[3] = 100
print(marks)

# slicing in lists - same as that of strings
print(marks[5:])
print(marks[:-2])
```

```python
# methods associated with lists
nums = [8,7,1,2]
nums.append(90) # add at the end of the list
print(nums)

nums.insert(3, 0) # add at a specific position
print(nums)

nums.sort() # sorts in ascending order
print(nums)

nums.sort(reverse=True) # sorts in descending order
print(nums)

nums.reverse() # reverses the given list
print(nums)
```

```python
# using loops with lists - we prefer for loops
nums = [1,2,3,90,4,5,10,6,7]

target = 6
ind = 0

for val in nums:
    if val == target:
        print(ind)
        break
    ind += 1
```

```python
# tuples - immutable sequence of values
# just like lists but they are immutable.
# tuples are created using ()
tup = (1,9,8,3,"hi",0,67,2)
print(type(tup))

print(tup[4])

# note - for creating a tuple with single value tup = (2) => this will be treated as an expression
# add a coma after that value to amke it tuple

tup2 = (2,)
print(type(tup2))

# slicing is also same as lists 
print(tup[2:7])
print(tup[:-1])
```

```python
# loops and functions with tuples
t = (1, 2, 3, 4)

sum = 0

for val in t:
    sum  += val

print(f"sum of values in the tuple : {sum}")
```

```python
# dictionaries - mutable unordered sequence of values
# duplicate keys are not allowed
dict = {
    "name" : "Shaik Khaja",
    "cgpa" : 9.4,
    "subjects" : ["math", "python", "c++"],
    3.14 : "PI"
}

print(dict)
print(type(dict))

# mutable
dict["cgpa"] = 9.9
print(dict)
```

```python
dict = {}
while True:
    print("\nA. Add Student")
    print("B. Update marks")
    print("C. Search student")
    print("D. Display all")
    print("Q. Quit")

    choice = input().upper()

    if choice == 'A':
        name = input("name?")
        marks = int(input("marks?"))
        dict.update({name : marks})
        print("student added")

    elif choice == "B":
        name = input("name?")
        if name in dict:
            marks = int(input("update marks?"))
            dict[name] = marks
            print("marks updated")
        else:
            print("student not found")

    elif choice == "C":
        name = input("name?")
        if name in dict:
            print(name, dict[name])
        else:
            print("student not found")

    elif choice == "D":
        print(dict.items())

    elif choice == "Q":
        print("Quiting...")
        break

    else:
        print("Invalid choice")
```

```python
l = [1,1,2,2,3,4,5,5,6]

seen = set()
duplicates = set()

for x in l:
    if x in seen:
        duplicates.add(x)
    else:
        seen.add(x)
    
print(duplicates)
```
